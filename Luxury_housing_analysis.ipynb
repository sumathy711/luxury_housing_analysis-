                                 1. filling null values in'Ticket_Price_Cr' column 

```python
#convert the str into numeric:
#errors='coerce' to handle the error if number in text

df['Ticket_Price_Cr'] =pd.to_numeric(df["Ticket_Price_Cr"],errors='coerce')

#to check the changes:
price_status=df['Ticket_Price_Cr'].describe()
print(price_status)
```

```python
df['Ticket_Price_Cr'].isnull().sum()
```

```python
import numpy as np
#1. replace negative values and 0 with NaN  
df.loc[df['Ticket_Price_Cr'] <=0, 'Ticket_Price_Cr'] =np.nan

#2. replace the null value with the median
price_median = df['Ticket_Price_Cr'].median()
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(price_median)
df['Ticket_Price_Cr'].describe()

```

```python
df['Ticket_Price_Cr'].isnull().sum()
```

                                  2.filling null values in 'Unit_Size_Sqft' column 

```python
#1. Turn negative/zero into NaN
df.loc[df['Unit_Size_Sqft'] <=0, 'Unit_Size_Sqft'] = np.nan
df['Unit_Size_Sqft'].describe()
```

```python
#fill the null value with the median value
size_median=df['Unit_Size_Sqft'].median()
df['Unit_Size_Sqft']=df['Unit_Size_Sqft'].fillna(size_median)
df['Unit_Size_Sqft'].describe()
```

```python
df['Unit_Size_Sqft'].isnull().sum()
```

                                       3.filling null values in 'Amenity_Score' column 

```python
df['Amenity_Score'].describe()
```

```python
df['Amenity_Score'].isnull().sum() 
```

```python
# filling the null values in 'Amenity_Score' column with mean:
df['Amenity_Score'] = df['Amenity_Score'].fillna(df['Amenity_Score'].mean())
df['Amenity_Score'].isnull().sum()
```

```python
df['Configuration'].isnull().sum()
```

```python
# filling the null values in 'Buyer_Comments ' column:
df['Buyer_Comments'] = df['Buyer_Comments'].fillna("No Comment Provided")
df['Buyer_Comments'].isnull().sum()
```

```python
# 'Configuration' column,'Micro_Market' column:
df['Configuration'] = df['Configuration'].str.upper().str.strip()
df['Micro_Market'] = df['Micro_Market'].str.title().str.strip()
```

```python
# 1. Clean the column names (Lowercase and strip spaces)
df.columns = df.columns.str.strip().str.lower()

# 2. Re-calculate using the NEW lowercase names
# (ticket_price_cr and unit_size_sqft)
df["price_per_sqft"] = (df["ticket_price_cr"] * 10000000) / df["unit_size_sqft"]

# 3. Handle the Date (purchase_quarter)
df['temp_date'] = pd.to_datetime(df['purchase_quarter'])
df['purchase_year'] = df['temp_date'].dt.year
df['quarter_number'] = df['temp_date'].dt.quarter

# 4. Create the Booking Flag
df['booking_flag'] = df['possession_status'].notna().astype(int)

# 5. Clean up and view
df.drop(columns=['temp_date'], inplace=True)
print(df[['purchase_quarter', 'price_per_sqft', 'purchase_year','booking_flag']].head())
```
